## VISUALLIZING WORLD DATA

```python
df_world_over_time = df_world.groupby(["date"])[["total_cases","new_cases","total_deaths","new_deaths"]].sum().reset_index()
df_world_over_time['date'] = pd.to_datetime(df_world_over_time['date']) #Change 'date' type from object to date
df_world_over_time
```

### Confirmed cases (Total Cases) all over the world

```python
plt.figure(figsize=(10,5))
plt.plot(df_world_over_time['date'],df_world_over_time['total_cases'] )
plt.xlabel('Months')
plt.ylabel('Confirmed cases')
plt.title('Confirmed cases (Total Cases) all over the world')
```

### Total deaths cases evolution over time

```python
plt.figure(figsize=(10,5))
plt.plot(df_world_over_time['date'],df_world_over_time['total_deaths'] )
plt.xlabel('Months')
plt.ylabel('Number of Deaths')
plt.title('Total deaths cases evolution over time')
```

### New cases all over the world

```python
plt.figure(figsize=(10,5))
plt.plot(df_world_over_time['date'],df_world_over_time['new_cases'], c='r' )
plt.xlabel('Months')
plt.ylabel('New Cases')
plt.title('Evolution of Covid-19 New Cases over time in the world')
```

### New Deaths all over the world

```python
plt.figure(figsize=(10,5))
plt.plot(df_world_over_time['date'],df_world_over_time['new_deaths'], c='k' )
plt.xlabel('Months')
plt.ylabel('New Deaths')
plt.title('Evolution of Covid-19 New Deaths over time in the world')
```

### Comparison between New Cases and New Deaths

```python
plt.figure(figsize=(10,5))
plt.plot(df_world_over_time['date'],df_world_over_time['new_cases'], c='r', label='New Cases' )
plt.plot(df_world_over_time['date'],df_world_over_time['new_deaths'], c='k', label='New Deaths')
plt.legend(loc=0)
plt.show()
```
